## 2. Clean and Standardize Column Names
To ensure consistency across the dataset, all column names will be converted to lowercase and stripped of leading/trailing spaces. This makes the DataFrame easier to work with and avoids errors during analysis.

```python
df.columns= df.columns.str.strip().str.lower()
df.columns
```

## 3. Clean Whitespace in Text Columns

To ensure consistency, all text (string/object) columns will be stripped of leading and trailing whitespace.  
This prevents issues during analysis, such as mismatches in categorical values caused by hidden spaces.

```python
for col in df.select_dtypes(include='object').columns:
    df[col]=df[col].astype(str).str.strip().str.lower()
df.head()

```

## 4. Text Standardization

To ensure categorical values are consistent, text fields will be standardized by:
- Converting all text to lowercase *(already completed)*.
- Mapping language codes (e.g., `en` → English, `ja` → Japanese).
- Removing extra spaces or punctuation inconsistencies *(done again after mapping to catch any new issues).*

This step ensures uniformity across columns such as `genre`, `director`, `cast`, and `language`.

```python
'''Standardize Language codes'''

unique_code = df["language"].unique()
```

```python
language_map = {"en" : "english" ,"ja": "japanese", "ko": "korean", "it": "italian",
    "fr": "french", "fa": "persian", "de": "german", "es": "spanish",
    "sv": "swedish", "pt": "portuguese", "cn": "chinese", "ru": "russian",
    "id": "indonesian", "zh": "chinese", "lv": "latvian", "he": "hebrew",
    "ro": "romanian", "ga": "irish", "te": "telugu", "da": "danish",
    "ar": "arabic", "tr": "turkish", "no": "norwegian", "pl": "polish",
    "hi": "hindi", "gl": "galician", "hu": "hungarian"}
unmapped_codes = [code for code in unique_code if code not in language_map]
print("Unmapped codes:\n" , unmapped_codes)

```

```python
df['language'] = df['language'].replace(language_map)
```

```python
df['language'].unique()
```

```python
''' Remove multiple spaces inside text'''
for col in df.select_dtypes(include='object').columns:
    df[col]=df[col].str.replace(r"\s+"," ",regex=True)
```

```python
'''Remove unwanted punctuation'''

import string

punctuation_to_remove = string.punctuation

for col in df.select_dtypes(include="object").columns:
    df[col]=df[col].str.replace(f"[{punctuation_to_remove}]","", regex=True)
```

```python
'''Remove whitespace again'''
for col in df.select_dtypes(include="object").columns:
    df[col]=df[col].str.strip()
```

## 5. Dropping Irrelevant Columns

Reminged irrelevant columns (`plot`, `awards`, `RT_URL`, `IMDb_10`) to retain consistency and make analysis cleaner and more focused.

```python
irrelevant_columns = ["streaming_on", "plot","awards","imdb_10","cast","director","production","rt_url"]
df = df.drop(columns=irrelevant_columns,errors="ignore")
```

```python
df.columns
```

## 6. Standardizing Missing Values

Converted inconsistent placeholders (like "NA", "null", "n/a") into `NaN` 
to ensure consistent handling of missing data across all columns.

```python
na_values = [ "na","nan"," ", "null"]
for value in df.select_dtypes(include="object").columns:
    df[value]=df[value].replace(na_values,np.nan)
```

## 7. Handling Missing Values
 
Numeric columns with missing values were imputed using the median to preserve distribution.  
Categorical/text columns were filled with `"Unknown"` to retain completeness without bias.
.

```python
'''Replacing missing values with Median'''
numerical_col=["metacritic", "audience_rating", 
                "audience_reviews", "letterboxd_votes", 
                "google_score"]
for col in numerical_col:
    df[col]=df[col].fillna(df[col].median())
```

```python
'''Replacing text missing values with unknown'''
for col in df.select_dtypes(include="object").columns:
    df[col]=df[col].fillna("unknown")
   
```

```python
df.isnull().sum()
```

## 8. Data Type Conversion

Converted `audience_reviews` and `letterboxd_votes` from float to intege 
to ensure consistency in numeric representation and simplify analysis.


```python
df['audience_reviews']=df['audience_reviews'].astype("int64")
df['letterboxd_votes']=df['letterboxd_votes'].astype("int64")
```

```python
df.dtypes
```

## 9. Quality Checks (before final verification)

Performed consistency checks to ensure related fields align (e.g., critic vs audience scores) and applied range validation to confirm numeric values fall within valid ranges (scores 0–100, years realistic).

```python
'''Consistency check: critic vs audience rating should not widely differ'''
score_diff = abs(df["critic_rating_rt"]-df["audience_rating"])
print(df[score_diff>50][["title","critic_rating_rt","audience_rating"]])
```

```python
''' Range validation: ensure scores are between 0 and 100 '''
for col in ["flickmetrix_score","imdb_100","metacritic","critic_rating_rt","audience_rating","google_score"]:
    invalid = df[(df[col]<0) | (df[col]>100)]
    if not invalid.empty:
        print(f"Invalid values in {col}:\n", invalid[[col]])
```

```python
''' Year validation'''
invalid_year= df[(df["year"]<1900) | (df["year"]>2026)]
print(invalid_year[["title","year"]])
```
